**Boxplot()** Kutu grafikleri, bir veri kumesinin dagilimini gosterir. Genellikle farkli kategorilerdeki bir veya daha fazla degiskenin dagilimini karsilastirmak icin kullanilirlar.

```python
import seaborn as sns
tips = sns.load_dataset("tips")  # tips veri seti
tips
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)  # gun ve toplam hesabi ifade eden kutu grafigi
plt.show()
```

Burada da goruldugu gibi kutu grafigi total bill degiskeni icin gunlere baktıgımızda persembe gunu dagılım 10-20 arasi ve 20 ye daha yakin cikarimi yapilabilir.

```python
# ozellestirme
sns.boxplot(
    x="day", 
    y="total_bill", 
    hue="time",    # burada ise hangi vakitlere gore oldugunun ayrimi da yapilmis
    data=tips, 
    palette="Set3",   # palette ="set3" renk paleti
    linewidth=1.5, 
    fliersize=4)

plt.title("Zamana ve Gunlere gore Toplam Hesap")

plt.xlabel("Gunler")

plt.ylabel("Toplam Hesap")

plt.show()
```

**Barplot()**  Cubuk grafikleri, kategorik bir degisken ile surekli bir degisken arasindaki iliskiyi gorsellestirmek icin kullanilir. Bir cubuk grafiginde, her cubuk her kategori icin surekli degiskenin ortalamasini veya medyanini (veya herhangi bir toplamayi) temsil eder.

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")  # titanic veri seti
titanic
```

```python
# cubuk grafik
sns.barplot(x="class",  # yolcu sinifina gore
            y="survived",  # hayatta kalma orani
            data=titanic)
plt.show()
```

```python
sns.barplot(x="class",
            y="survived",
            hue="sex",  # sex e gore gruplandirilmis
            palette="muted", # renk paleti
            data=titanic)

plt.xlabel("class")
plt.ylabel("hayatta kalma orani")
plt.title("Cinsiyete gore hayatta kalma orani")
plt.show()
```

<h1>Dagilim Grafikleri</h1>

Dagilim grafikleri, iki surekli degisken arasindaki iliskiyi gorsellestirmek icin kullanilir.

**Scatterplot()**

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips
```

```python
sns.scatterplot(x="total_bill",  # toplam hesap
                y="tip",   # bahsis
                data=tips)
plt.show()
```

```python

sns.scatterplot(x="total_bill",
                y="tip",
                hue="time",   # time a gore renk gruplandirmasi
                size="size",  # masa boyutu 
                palette="magma",
                sizes=(50, 200), data=tips)  # en buyuk masa 200,en kucuk 50 

plt.xlabel("Toplam Hesap")

plt.ylabel("Bahsis")

plt.title("Toplam hesap ve Bahsis arasindaki iliski")
```

**Jointplot()**

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips
```

```python
sns.jointplot(x="total_bill", 
              y="tip", 
              kind="hex",   # grafikte goruntulenecek sekil
              cmap="flare",  # renk
              data=tips)

plt.show()
```

**Heatmap()**

Isı haritası, matrisin degerini gorsellestirmek icin renkler kullanan verilerin grafiksel
bir gosterimi olarak tanimlanir.Daha yaygin degerleri temsil etmek icin temel olarak daha parlak renkler kırmızımsı renkler kullanilir ve daha az yaygin değerlerini temsil etmek icin daha koyu renkler tercih edilir.

```python
penguins = sns.load_dataset("penguins")
num_data = penguins.select_dtypes(include=['float64', 'int64'])
corr_data = num_data.corr()   # sayisal veriler arasi iliski
corr_data
```

```python
sns.heatmap(corr_data,
           annot=True,
           cmap="crest")
```
