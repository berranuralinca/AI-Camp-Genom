**Odev 1: Matplotlib Uygulamasi**

**Grafik Cizimi**

```python
# Bir cizgi grafigi olustur ve eksenlere baslik ekle.
# Ayni  grafikte bir cubuk grafigi goster.
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array(["Ali","Veli","Ayse"])  # isim bilgisi 
y = np.array([180,175,160])  # boy bilgisi
z = np.array([55,90,80])   # kilo bilgisi

fig,ax = plt.subplots()  # grafikleri birlestirmek

# cizgi grafigi ax.plot(isim,boy,color,label,linewidth,marker,markersize)
ax.plot(x,z,ls=":",color="red",label="kilo",lw=2.0,marker=".",markersize=10.0)  
# cubuk grafigi ax.bar(isim,kilo,widthl)
ax.bar(x,y,width=0.1,color="grey")

ax.set_xlabel('isimler') # x ekseni basligi
ax.set_ylabel('boy')   # y ekseni basligi
ax.set_title('Ogrenciler')  # baslik
ax.legend() # g√∂sterge
plt.show()  # grafigi goster
```

**Alt Grafikler**

- Tek bir figur icinde iki farkli alt grafik olusturma

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)  # Seed degeri ayni oldugu surece, ayni rastgele sayi dizisi
x = np.random.rand(100)  # rastgele 100 x degeri
y = 2 * x + np.random.normal(0, 0.1, 100)  # y = 2x + g√ºr√ºlt√º dogru etrafinda dagilmis veriler

# fig,(ilk grafik,ikinci grafik)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))    # direkt iki ayri grafik cizip subplot() ile de birlestirebilirdik.

# ƒ∞lk grafik, Scatter plot 
axes[0].scatter(x, y, color='blue', alpha=0.7)  # ax1.scatter(x,y,color,seffaflik)
axes[0].set_xlabel('time')    # zaman a gore virus miktari
axes[0].set_ylabel('amount of virus')
axes[0].set_title('Scatter Plot')

# ƒ∞kinci grafik, Histogram
axes[1].hist(y, color='green', alpha=0.7)  #ax2.hist(y,color,seffaflik)
axes[1].set_xlabel('amount of virus')
axes[1].set_ylabel('frekans')
axes[1].set_title('Histogram')

# Grafiƒüi g√∂sterme
plt.show()
```

**Grafik Kaydetme**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

colors = np.random.randint(100, size=(100))  # rastgele renklendirme
sizes = 10 * np.random.randint(100, size=(100))   # rastgele boyutlandirma

#dagƒ±lƒ±m grafigi 
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral') #renk haritasi kullanma

plt.show()
plt.savefig("colors.png")  # grafik kaydetme
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")
penguins  # pengunler veri seti
```

```python
# bu scatterplot gaga uzunlugu ve gaga derinligi arasindaki iliskiyi ture gore belirler.
sns.scatterplot(x="bill_length_mm", # gaga uzunlugu
                y="bill_depth_mm", # gaga derinligi
                hue="species",  # tur
                data=penguins,
                palette="magma")  # color
plt.show()
```

```python
# ayni veriyle displot
```

```python
sns.displot(data=penguins, x="bill_length_mm",hue="species",kde=True,palette="viridis")
plt.show()
```

```python
sns.kdeplot(data=penguins, 
            x="bill_length_mm",   # gaga uzunlugunun turlere gore yogunlugu
            hue="species",
            palette="viridis",
            fill=True)  # golge ekleme
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips  # tips veri seti
```

```python
sayisal = tips.select_dtypes(include=['float64', 'int64'])
kordata = sayisal.corr()   # sayisal veriler arasi iliski
kordata
```

```python
from matplotlib.colors import ListedColormap

renkler = [ "purple", "blue","pink"] # renk skalasi ozellestirme
cmap = ListedColormap(renkler)
sns.heatmap(kordata,  
           annot=True,   # hucre icinde degerler
            fmt=".1f",  # basamak sayisi
           cmap=cmap)
```
