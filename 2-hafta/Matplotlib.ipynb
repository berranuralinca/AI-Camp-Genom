**Marker Size**

"markersize" ya da "ms" anahtar kelimesi kullanilir.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,"o:",ms = 7)
plt.show()
```

**Labels and Title**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([1, 2, 3, 4])

plt.plot(x, y)
plt.xlabel("x eksenim")   # x ekseni icin baslik
plt.ylabel("y eksenim")   # y ekseni icin baslik
plt.title("Grafik 1")     # grafik icin baslik ekleme
plt.show()
```

**Line and Title Styles**

- font1 ={"family":"serif","color":"red","size":10}  baslik icin
- font2 ={"family":"serif","color":"green","size":5} label icin

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([1, 2, 3, 4])

font1 ={"family":"serif","color":"red","size":10}  
font2 ={"family":"serif","color":"green","size":8} 

plt.plot(x, y)
plt.xlabel("x eksenim",fontdict=font2)   # x ekseni icin baslik
plt.ylabel("y eksenim",fontdict=font2)   # y ekseni icin baslik
plt.title("Grafik 1",fontdict=font1,loc="left")     # grafik icin baslik ekleme
plt.show()
```

- basligi konumlandirmak icin loc:location
- left,right,center

plt.scatter(x, y, s=sizes)  Noktalarin boyutunu degistirmek icin kullanilir.

```python
import matplotlib.pyplot as plt
import numpy as np
# opak grafik
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])  # farkli nokta boyutu vermek icin 
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.subplot(1,2,1)
plt.scatter(x, y,c=colors, s=sizes,cmap="CMRmap")  # boyut ayarlama

# seffaf grafik

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])  # farkli nokta boyutu vermek icin 
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.subplot(1,2,2)
plt.scatter(x, y,c=colors, s=sizes,cmap="CMRmap",alpha=0.7)  # seffaflik ayarlama

plt.show()
```

**Cubuk Grafik**

**Annotate()**

Matplotlib kutuphanesinin pyplot modulundeki annotate() fonksiyonu, xy noktasini s metniyle aciklama eklemek icin kullanilir .

plt.annotate(metin,konum, xytext=metnin konumu, xycoords='data', textcoords='data', arrowprops=ok ozelligi)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker='o', linestyle='-', color='blue')

plt.annotate('Peak', xy=(3, 5), xytext=(3.5, 6), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Annotated Plot')
plt.grid(True)
plt.show()
```
