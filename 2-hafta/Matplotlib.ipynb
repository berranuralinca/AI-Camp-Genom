**Markers**

Markers >> Her noktayi isaretci ile belirlemek.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,marker = "o")      # noktalarla belirtme
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,marker = "*")      # yildizla belirtme
plt.show()
```

**Format Strings fmt**

- marker: isaretleyici
- line: satir veya cizgi
- color: renk

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,"o:r")      # "o:isaretci , : cizgi sekli ,r:renk"
plt.show()
```

**Line Referance** :

"linestyle" ya da "ls"
- solid line "-"
- dotted line ":"
- dashed line "--"
- dashed/dotted line "-."

**Color Referance**

- red "r"
- green "g"
- blue "b"
- cyan "c"
- magenta "m"
- yellow "y"
- black "k"
- white "w"

**Marker Size**

"markersize" ya da "ms" anahtar kelimesi kullanilir.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,"o:",ms = 7)
plt.show()
```

**Marker Color**

- "markeredgecolor" ya da "mec" kenar rengi
- "markerfacecolor" ya da "mfc" ic rengi

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,"o:",ms = 7,mec="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,"o:",ms = 7,mec="r",mfc="r")
plt.show()
```

**Line Color**

"color" ya da "c"

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,":",c ="b")
plt.show()
```

**Line Width**

"linewidth" ya da "lw"

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,c ="r",lw=3.0)  # line width float deger alir.
plt.show()
```

**Labels and Title**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([1, 2, 3, 4])

plt.plot(x, y)
plt.xlabel("x eksenim")   # x ekseni icin baslik
plt.ylabel("y eksenim")   # y ekseni icin baslik
plt.title("Grafik 1")     # grafik icin baslik ekleme
plt.show()
```

**Line and Title Styles**

- font1 ={"family":"serif","color":"red","size":10}  baslik icin
- font2 ={"family":"serif","color":"green","size":5} label icin

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([1, 2, 3, 4])

font1 ={"family":"serif","color":"red","size":10}  
font2 ={"family":"serif","color":"green","size":8} 

plt.plot(x, y)
plt.xlabel("x eksenim",fontdict=font2)   # x ekseni icin baslik
plt.ylabel("y eksenim",fontdict=font2)   # y ekseni icin baslik
plt.title("Grafik 1",fontdict=font1,loc="left")     # grafik icin baslik ekleme
plt.show()
```

- basligi konumlandirmak icin loc:location
- left,right,center

**Matplotlib Grid**

**Bir Grafige Sutun Cizgisi Ekleme**

Pyplot ile grid() grafige sutun cizgisi eklememizi saglar.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,90,100,110])
y = np.array([250,300,350,400])
plt.plot(x,y)
plt.title("Sports Watch Data")
plt.xlabel("average pulse")
plt.ylabel("calorie burnage")
plt.grid()    # sutun ekleme
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,90,100,110])
y = np.array([250,300,350,400])
plt.plot(x,y)
plt.title("Sports Watch Data")
plt.xlabel("average pulse")
plt.ylabel("calorie burnage")

# hangi sutun cizgilerinin gosterilecegini axis ile belirleme
plt.grid(axis="x")     # x de gosterme
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,90,100,110])
y = np.array([250,300,350,400])
plt.plot(x,y)
plt.title("Sports Watch Data")
plt.xlabel("average pulse")
plt.ylabel("calorie burnage")

# hangi sutun cizgilerinin gosterilecegini axis ile belirleme
plt.grid(axis="y")   # y de gosterme
plt.show()
```

**Subplots**

**Coklu Grafik Goruntuleme**

Tek bir yerde birden fazla grafik cizmeyi saglar.

```python
# plt.subplot(1,2,1)   1 satir 2 sutundan olusan graffikte ilk grafik
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,90,100,110])
y = np.array([250,300,350,400])

plt.subplot(1,2,1)   # coklu grafik cizme
plt.plot(x,y)
plt.title("sol")

x = np.array([10,20,30])
y = np.array([1,2,3])
plt.subplot(1,2,2)      # yan yana cizme
plt.plot(x,y)
plt.title("sag")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,90,100,110])
y = np.array([250,300,350,400])

plt.subplot(2,1,1)   # alt alta cizme
plt.plot(x,y)
plt.title("alt")

x = np.array([10,20,30])
y = np.array([1,2,3])
plt.subplot(2,1,2)
plt.plot(x,y)
plt.title("ust")
plt.show()
```

```python
# daha fazla grafik cizme

import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,90,100,110])
y = np.array([250,300,350,400])
plt.subplot(1,3,1)   # coklu grafik cizme
plt.plot(x,y)
plt.title("ilk grafik")

x = np.array([10,20,30])
y = np.array([1,2,3])
plt.subplot(1,3,2)
plt.plot(x,y)
plt.title("ikinci grafik")


x = np.array([10,20,30])
y = np.array([1,2,3])
plt.subplot(1,3,3)
plt.plot(x,y)
plt.title("ucuncu grafik")

plt.suptitle("grafikler")    # ana baslik gibi dusunebiliriz.
plt.show()
```
