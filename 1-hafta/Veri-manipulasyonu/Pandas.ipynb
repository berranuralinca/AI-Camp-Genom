**Pandas Serisi Olusturmak**

```python
import pandas as pd

NewSeries = pd.Series([1,2,3,4,5])
print(NewSeries)

print(type(NewSeries))
print(NewSeries.axes)
print(NewSeries.dtype)
print(NewSeries.size)
print(NewSeries.ndim)

print(NewSeries.values)
print(NewSeries.head(3))
print(NewSeries.tail(2))

NewSeries = pd.Series([2,3,4,5,6],index = ["a","b","c","d","e"])
print(NewSeries)
print(NewSeries["c"])

# Sozluk uzerinden
Dict = {"abc":123,"def":456,"ghk":789}
DictSeries = pd.Series(Dict)
print(DictSeries)
print(DictSeries["abc"])
```

**Birlestirme**

```python
import pandas as pd
import numpy as np

Arr = np.random.randint(1,10,(7,4))
df = pd.DataFrame(Arr,columns = ["var1","var2","var3","var4"])
df

df2 = df + 10
df2
```

```python
pd.concat([df,df2],ignore_index = True)       # indeksi devam ettirir.
```

```python
df2.columns = ["var1","var2","deg1","deg2"]
```

```python
pd.concat([df,df2],join ="inner")       # kesisimlerini birlestirir
```

**Toplulastirma ve Gruplama**

```python
# count() ,first() ,last() ,mean() ,median() ,min() ,max() ,std() , var() ,sum()

import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset("planets")
df.head()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"gruplar":["kiz","erkek","kiz","kiz"],"yas":[10,11,9,5]})
df
```

```python
df.groupby("gruplar").mean()    # gruplara göre yas ortalamasi
```

```python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset("Planets")
df.head()
```

```python
df.dropna().groupby("method")["mass"].mean()
```

```python
df.groupby("method").aggregate({"mass":"mean","year":"max"})
```
