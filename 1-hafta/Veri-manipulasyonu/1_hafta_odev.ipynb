**Matris Olusturma ve Islemler**
 3x3 boyutunda rastgele tamsayilardan olusan bir NumPy matrisi olusturun.
Bu matrisin:
- Transpozesini alin.
- Determinantini hesaplayin.
- Her elemanini 2 ile carpan bir matris olusturun

```python
import numpy as np
RandIntArr = np.random.randint(0,100,(3,3))
print(RandIntArr)

TransArr = RandIntArr.T
print(TransArr)

Det = np.linalg.det(RandIntArr)
print(Det)

MultiArr = RandIntArr * 2
print(MultiArr)
```

##### **Dizi Operasyonlari**
1-50 arasinda 10 esit aralikli sayi iceren bir numpy dizisi olusturun.Bu dizinin:

- Ortalamasini ve standart sapmasini hesaplayin.
- Elemanlarinin karesini alin.
- Elemanlardan 25 'ten buyuk olanları filtreleyip ekrana yazdirin.

```python
import numpy as np

ArangeArr = np.arange(1,50,6)
print(ArangeArr)

Ortalama = np.mean(ArangeArr)
print(Ortalama)

StandSap = np.std(ArangeArr)
print(StandSap)

SquareArr =(ArangeArr ** 2)
print(SquareArr)

Condition = ArangeArr > 25
FilteredArr = ArangeArr[Condition]
print(FilteredArr)
```

**Numpy Fonksiyonlari ile Calisma**
Sirasiyla sinus ve kosinus degerlerini hesaplamak icin 0 ile 2pi arasinda 100 deger üreten bir numpy dizisi olusturun.

- Sin ve cos değerlerini ayni grafikte cizdirin.

```python
import numpy as np

TriArr = np.linspace(0,2*np.pi,100)
print(TriArr)

SinArr = np.sin(TriArr)
print(SinArr)
CosArr = np.cos(TriArr)
print(CosArr)

from matplotlib import pyplot as plt
plt.plot(TriArr,SinArr,"b-", label="sinus")
plt.plot(TriArr,CosArr,"r-", label="kosinus")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.show()
```

```python
# Hayatta kalma oranini hesaplayin.
file_path = "C:/Users/berra/OneDrive/Masaüstü/Titanic-Dataset.csv"
df = pd.read_csv(file_path,sep = ",")  

HayattaKalma = df.groupby("Sex")["Survived"].sum() / df.groupby("Sex")["PassengerId"].count()
print(HayattaKalma)
```

```python
# Yolculari sinif sutununa (Pclass) gore gruplayarak hayatta kalma oranlarini hesaplayin.

Pclass_df = df.groupby("Pclass")["Survived"].mean()
```

```python
# Bu gruplara ait hayatta kalma oranlarini bir cubuk grafikle gorsellestirin.

import matplotlib.pyplot as plt

Pclass_df.plot(kind = "bar")         # cubuk grafigi
plt.xlabel("passenger class")
plt.ylabel("Survived rate")
plt.show()
```
